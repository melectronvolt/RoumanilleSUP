```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function f(x) = cos(3x+2)
def f(x):
    return np.cos(3*x + 2)

# Numerical derivative of f(x) using the method (f(a+h)-f(a))/h
def numerical_derivative(f, a, h=1e-5):
    return (f(a+h) - f(a)) / h

# Analytical derivative of f(x), which is -3sin(3x+2)
def analytical_derivative(x):
    return -3 * np.sin(3*x + 2)
```

```python
# Create an array of x values from -2pi to 2pi
x = np.linspace(-2*np.pi, 2*np.pi, 400)

# Calculate function and derivative values
y_f = f(x)
y_numerical_derivative = numerical_derivative(f, x)
y_analytical_derivative = analytical_derivative(x)

# Plotting
plt.figure(figsize=(10, 6))

# Plot original function
plt.plot(x, y_f, label='f(x) = cos(3x+2)')

plt.title("Function and Its Derivatives")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))

# Plot original function
plt.plot(x, y_f, label='f(x) = cos(3x+2)')


# Plot numerical derivative
plt.plot(x, y_numerical_derivative, label="Numerical Derivative of f(x)", linestyle='--')
plt.title("Function and Its Derivatives")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting
plt.figure(figsize=(10, 6))

# Plot original function
plt.plot(x, y_f, label='f(x) = cos(3x+2)')

# Plot numerical derivative
plt.plot(x, y_numerical_derivative, label="Numerical Derivative of f(x)", linestyle='--')

# Plot analytical derivative
plt.plot(x, y_analytical_derivative, label='Analytical Derivative: -3sin(3x+2)', linestyle=':')
plt.title("Function and Its Derivatives")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()



```
